```python
import pandas as pd

df = pd.DataFrame({
    'poeta': ['Eliot', 'Frost', 'Milton'],
    'obras': [12, 20, 15]
})

# Criar uma nova coluna com uma classificação baseada no número de obras
def classificar_produtividade(obras):
    if obras >= 20:
        return 'Alta'
    elif obras >= 15:
        return 'Média'
    else: 
        return 'Baixa'

df['produtividade'] = df['obras'].apply(classificar_produtividade)
df
```

```python
df['descrição'] = df.apply(lambda linha: f'{linha['poeta']} escreveu {linha['obras']} obras', axis=1)
df
```

```python
df[['poeta', 'descrição']].map(lambda x: x.upper() if isinstance(x, str) else x)
df
```

```python
import numpy as np
import pandas as pd

df = pd.DataFrame({
    'poeta': ['Eliot', 'Frost', 'Milton', 'Wordsworth'],
    'obras': [12, 20, 15, 10]
})

df['classificação'] = np.where(
    df['obras'] >= 20, 'Muito produtivo',
    np.where(df['obras'] >= 15, 'Produtivo', 'Pouco produtivo')
)

df
```

```python
df = pd.DataFrame({
    'poeta': ['Eliot', 'Frost', 'Milton', 'Wordsworth'],
    'idade': [45, 60, 38, 70]
})

faixas = [0, 40, 60, 100]
rotulos = ['Jovem', 'Adulto', 'Idoso']

df['faixa_etaria'] = pd.cut(df['idade'], bins=faixas, labels=rotulos)

df
```

```python
import pandas as pd

# Exemplo com dados de idade

df = pd.DataFrame({
    'poeta': ['Eliot', 'Frost', 'Milton', 'Wordsworth', 'Desconhecido'],
    'idade': [45, 50, 38, 70, 200] # 200 é um outlier
})

# Cálculo dos quartis

q1 = df['idade'].quantile(0.25)
q3 = df['idade'].quantile(0.75)
iqr = q3 - q1

# Limites
limite_inferior = q1 - 1.5 * iqr
limite_superior = q3 + 1.5 * iqr

# Substituir outliers por NaN

df['idade_ajustada'] = df['idade'].mask(
    (df['idade'] < limite_inferior) | (df['idade'] > limite_superior)
)

df
```

```python
import pandas as pd

df = pd.DataFrame({
    'poet': ['Eliot', 'Frost', 'Milton', 'Eliot', 'Frost'],
    'poem': ['Four Quartets', 'The Road Not Taken', 'Paradise Lost', 'The Waste Land', 'Fire and Ice'],
    'quantity': [5, 2, 7, 3, 1]
})

# Agrupar por poeta e somar a quantidade de obras
agrupado = df.groupby('poet')['quantity'].sum()
df
```

```python
from IPython.display import display

# Basic descriptive statistics for all numeric columns
display(df.describe())

# Descriptive statistics for the 'quantity' column
print(f'Mean: {df['quantity'].mean()}')
print(f'Median: {df['quantity'].median()}')
print(f'Std: {df['quantity'].std()}') # Standard deviation

# Grouped statistics by poet
print("\nMean quantity per poet:")
display(df.groupby('poet')['quantity'].mean())
```

```python
import pandas as pd

# First table: authors
authors = pd.DataFrame({
    'poet': ['Eliot', 'Frost', 'Milton'],
    'birth_year': [1888, 1874, 1608]
})

# Second table: works
works = pd.DataFrame({
    'poet': ['Eliot', 'Frost', 'Milton', 'Wordsworth'],
    'work': ['Four Quartets', 'The Road Not Taken', 'Paradise Lost', 'The Prelude']
})

# Merge on the 'poet' column
merged = pd.merge(authors, works, on='poet', how='inner') # Only matches
display(merged)
```
